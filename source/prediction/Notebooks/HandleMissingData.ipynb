```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/csv_data_march_16.csv', low_memory = False)
```

```python
pd.set_option('display.max_columns', 50)
df.columns = [c.replace(' ', '_') for c in df.columns]
```

```python
akarsu_df = df.loc[df['Table_Type'] == 'Akarsu']
gol_df = df.loc[df['Table_Type'] == 'Göl']
aritma_df = df.loc[df['Table_Type'] == 'Arıtma']
deniz_df = df.loc[df['Table_Type'] == 'Deniz']
```

```python
# Display missing data
missing = df.isna().sum()
missing_perc = missing / len(df)
m = pd.DataFrame(missing)
m_perc = pd.DataFrame(missing_perc)
m_df = pd.concat([m,m_perc], axis = 1)
m_df = m_df.rename(columns = {0: 'Total'})
# Create akarsu dataframe that has missing info
missing_akarsu = akarsu_df.isna().sum()
missing_akarsu_perc = missing_akarsu / len(akarsu_df)
m_akarsu = pd.DataFrame(missing_akarsu)
m_akarsu_perc = pd.DataFrame(missing_akarsu_perc)
m_akarsu_df = pd.concat([m_akarsu,m_akarsu_perc], axis = 1)
m_akarsu_df = m_akarsu_df.rename(columns = {0: 'Akarsu'})
# Create göl dataframe that has missing info
missing_gol = gol_df.isna().sum()
missing_gol_perc = missing_gol / len(gol_df)
m_gol = pd.DataFrame(missing_gol)
m_gol_perc = pd.DataFrame(missing_gol_perc)
m_gol_df = pd.concat([m_gol,m_gol_perc], axis = 1)
m_gol_df = m_gol_df.rename(columns = {0: 'Göl'})
# Create arıtma dataframe that has missing info
missing_aritma = aritma_df.isna().sum()
missing_aritma_perc = missing_aritma / len(aritma_df)
m_aritma = pd.DataFrame(missing_aritma)
m_aritma_perc = pd.DataFrame(missing_aritma_perc)
m_aritma_df = pd.concat([m_aritma,m_aritma_perc], axis = 1)
m_aritma_df = m_aritma_df.rename(columns = {0: 'Arıtma'})
# Create deniz dataframe that has missing info
missing_deniz = deniz_df.isna().sum()
missing_deniz_perc = missing_deniz / len(deniz_df)
m_deniz = pd.DataFrame(missing_deniz)
m_deniz_perc = pd.DataFrame(missing_deniz_perc)
m_deniz_df = pd.concat([m_deniz,m_deniz_perc], axis = 1)
m_deniz_df = m_deniz_df.rename(columns = {0: 'Deniz'})
# Merge into one dataframe
m_all = pd.concat([m_df, m_akarsu_df, m_gol_df, m_aritma_df,m_deniz_df], axis = 1)
m_all
```

```python
# Akarsu -> Toplam pestisid, cozunmus oksijen, KOI, Toplam Kjedahl azotu, Nitrat azotu, toplam fosfor, fekal koliform, toplam koliform
rel_col_names_akarsu = ['Numune_Adı', 'Tarih', 'Fekal_Koliform', 'Toplam_Koliform', 'Toplam_Fosfor', 'Toplam_Kjeldahl_Azotu', 'Toplam_Pestisit', 'Kimyasal_Oksijen_İhtiyacı', 'Nitrat_Azotu', 'Çözünmüş_Oksijen']
akarsu_rel = akarsu_df[rel_col_names_akarsu]
# Göl -> Toplam fosfor, toplam azot, klorofil-a
rel_col_names_gol = ['Numune_Adı', 'Tarih', 'Toplam_Fosfor', 'Toplam_Azot', 'Klorofil']
gol_rel = gol_df[rel_col_names_gol]
# Deniz -> Toplam koliform, fekal koliform, fekal streptokok, amonyak
rel_col_names_deniz = ['Numune_Adı', 'Tarih', 'Toplam_Koliform', 'Fekal_Koliform', 'Amonyak', 'Fekal_Streptokok']
deniz_rel = deniz_df[rel_col_names_deniz]
# Arıtma -> BOİ,KOİ, toplam fosfor, toplam azot
rel_col_names_aritma = ['Numune_Adı', 'Tarih', 'Biyokimyasal_Oksijen_İhtiyacı', 'Kimyasal_Oksijen_İhtiyacı', 'Toplam_Fosfor', 'Toplam_Azot']
aritma_rel = aritma_df[rel_col_names_aritma]
```

```python
# Missing relevant features for Akarsu
missingr_akarsu = akarsu_rel.isna().sum()
missingr_akarsu_perc = missingr_akarsu / len(akarsu_rel)
mr_akarsu = pd.DataFrame(missingr_akarsu)
mr_akarsu_perc = pd.DataFrame(missingr_akarsu_perc)
mr_akarsu_df = pd.concat([mr_akarsu,mr_akarsu_perc], axis = 1)
mr_akarsu_df = mr_akarsu_df.rename(columns = {0: 'Akarsu'})
mr_akarsu_df
```

```python
# Missing relevant features for Göl
missingr_gol = gol_rel.isna().sum()
missingr_gol_perc = missingr_gol / len(gol_rel)
mr_gol = pd.DataFrame(missingr_gol)
mr_gol_perc = pd.DataFrame(missingr_gol_perc)
mr_gol_df = pd.concat([mr_gol,mr_gol_perc], axis = 1)
mr_gol_df = mr_gol_df.rename(columns = {0: 'Göl'})
mr_gol_df
```

```python
# Missing relevant features for Deniz
missingr_deniz = deniz_rel.isna().sum()
missingr_deniz_perc = missingr_deniz / len(deniz_rel)
mr_deniz = pd.DataFrame(missingr_deniz)
mr_deniz_perc = pd.DataFrame(missingr_deniz_perc)
mr_deniz_df = pd.concat([mr_deniz,mr_deniz_perc], axis = 1)
mr_deniz_df = mr_deniz_df.rename(columns = {0: 'Deniz'})
mr_deniz_df
```

```python
# Missing relevant features for Arıtma
missingr_aritma = aritma_rel.isna().sum()
missingr_aritma_perc = missingr_aritma / len(aritma_rel)
mr_aritma = pd.DataFrame(missingr_aritma)
mr_aritma_perc = pd.DataFrame(missingr_aritma_perc)
mr_aritma_df = pd.concat([mr_aritma, mr_aritma_perc], axis = 1)
mr_aritma_df = mr_aritma_df.rename(columns = {0: 'Arıtma'})
mr_aritma_df
```

```python
# Converting all dataframes to time series.
ts_akarsu = akarsu_rel.set_index('Tarih')
ts_akarsu.to_csv("../RelevantData/akarsu_relevant.csv")

ts_gol = gol_rel.set_index('Tarih')
ts_gol.to_csv("../RelevantData/gol_relevant.csv")

ts_deniz = deniz_rel.set_index('Tarih')
ts_deniz.to_csv("../RelevantData/deniz_relevant.csv")

ts_aritma = aritma_rel.set_index('Tarih')
ts_aritma.to_csv("../RelevantData/aritma_relevant.csv")
```
